```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\tejas\Downloads\Dataset .csv")
```

ðŸ”¹ LEVEL 2 â€“ TASK 1: Restaurant Ratings

ðŸŽ¯ Objectives

Distribution of aggregate ratings

Most common rating range

Average number of votes

```python
# Rating distribution
plt.figure(figsize=(6,4))
sns.histplot(df['Aggregate rating'], bins=10, kde=True)
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Most common rating range
common_rating = df['Aggregate rating'].value_counts().idxmax()

# Average votes
avg_votes = df['Votes'].mean()

print("Most common rating:", common_rating)
print("Average number of votes:", round(avg_votes, 2))
```

ðŸ“˜ Conclusion 

1) Most restaurants fall between 3.0 â€“ 4.0 ratings
2) Average votes indicate moderate customer engagement

ðŸ”¹ LEVEL 2 â€“ TASK 2: Cuisine Combination

ðŸŽ¯ Objectives

Most common cuisine combinations

Check if combinations have higher ratings

```python
# Count cuisine combinations
top_combinations = df['Cuisines'].value_counts().head(10)
print("Top Cuisine Combinations:")
print(top_combinations)

# Average rating by cuisine combination
avg_rating_combo = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print("\nTop Rated Cuisine Combinations:")
print(avg_rating_combo)
```

ðŸ“˜ Conclusion

1) Certain cuisine combinations (e.g., North Indian + Chinese)
    tend to have higher average ratings
2) Popularity does not always mean higher rating

ðŸ”¹ LEVEL 2 â€“ TASK 3: Geographic Analysis

ðŸŽ¯ Objectives

Plot restaurants using latitude & longitude

Identify clusters

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.title("Geographic Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
plt.figure(figsize=(6,6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.title("Geographic Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

ðŸ“˜ Conclusion

1) Restaurants are densely clustered in metropolitan areas
2) Sparse distribution in rural regions

ðŸ”¹LEVEL 2 â€“ TASK 4: Restaurant Chains

ðŸŽ¯ Objectives

Identify restaurant chains

Analyze ratings & popularity

```python
# Identify chains (restaurants with multiple outlets)
chains = df['Restaurant Name'].value_counts()
restaurant_chains = chains[chains > 1]

print("Top Restaurant Chains:")
print(restaurant_chains.head())

# Ratings of chains
chain_ratings = df.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print("\nTop Rated Restaurant Chains:")
print(chain_ratings)
```

ðŸ“˜ Conclusion

2) Chain restaurants generally receive more votes
1) Ratings are consistent across outlets, indicating brand trust
