ðŸ”¹ Task 1: Top Cuisines

What to do

Find top 3 cuisines

Calculate percentage of restaurants serving them

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\tejas\Downloads\Dataset .csv")

# Split multiple cuisines
cuisines = df['Cuisines'].dropna().str.split(', ')

# Count cuisines
all_cuisines = cuisines.explode()
top_3 = all_cuisines.value_counts().head(3)

# Percentage
percentage = (top_3 / len(df)) * 100

print("Top 3 Cuisines:")
print(top_3)
print("\nPercentage:")
print(percentage)
```

ðŸ”¹ Task 2: City Analysis

What to do

City with most restaurants

Average rating per city

City with highest average rating

```python
# City with highest number of restaurants
top_city = df['City'].value_counts().idxmax()

# Average rating per city
avg_rating_city = df.groupby('City')['Aggregate rating'].mean()

# City with highest average rating
best_rated_city = avg_rating_city.idxmax()

print("City with most restaurants:", top_city)
print("City with highest avg rating:", best_rated_city)
```

ðŸ”¹ Task 3: Price Range Distribution

What to do

Bar chart

Percentage of each price range

```python
import matplotlib.pyplot as plt

price_counts = df['Price range'].value_counts()

# Plot
price_counts.plot(kind='bar')
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()

# Percentage
price_percentage = (price_counts / len(df)) * 100
print(price_percentage)
```

ðŸ”¹ Task 4: Online Delivery

What to do

% of restaurants offering online delivery

Compare ratings

```python
# Calculate percentage of online delivery availability
online_percentage = (
    df['Has Online delivery']
    .value_counts(normalize=True)
    * 100
)

# Calculate average rating based on online delivery
ratings_compare = (
    df.groupby('Has Online delivery')['Aggregate rating']
    .mean()
)

print("Online Delivery Percentage:")
print(online_percentage.round(2))

print("\nAverage Ratings:")
print(ratings_compare.round(2))
```
